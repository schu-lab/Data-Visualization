```python
# Author: Simon Chu
# Date: 2021-11-07

# Read World Happiness Data from 2017
# https://raw.githubusercontent.com/ebubekiraygun/World-Happiness-Data-on-Python/master/2017.csv

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ebubekiraygun/World-Happiness-Data-on-Python/master/2017.csv')
```

```python
df.head()
```

```python
# Slicing for GDP.per.Capita vs. Happiness.Score
df[['Happiness.Score', 'Economy..GDP.per.Capita.']].head()
```

```python
# Slicing Happiness
Happiness = df[['Happiness.Score']]
```

```python
# Slicing Economy
GDP = df[['Economy..GDP.per.Capita.']]
```

```python
# Plot for GDP.per.Capita vs. Happiness.Score
plt.title("GDP.per.Capita vs. Happiness.Score")
plt.xlabel('GDP')
plt.ylabel('Happiness Score')
plt.scatter(GDP, Happiness, label = 'GDP', color = 'blue')
plt.legend()
```

```python
# Slicing Freedom
Freedom = df[['Freedom']]
```

```python
# Plot for Freedom vs. Happiness.Score
plt.title("Freedom vs. Happiness.Score")
plt.xlabel('Freedom')
plt.ylabel('Happiness Score')
plt.scatter(Freedom, Happiness, label = 'Freedom', color = 'green')
plt.legend()
```

```python
#Correlate Heatmap Matrix
df.corr()
```
